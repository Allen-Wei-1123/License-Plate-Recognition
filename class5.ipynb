```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('cat.jpeg',0);
hist = cv2.calcHist([img],[0],None,[256],[0,256])
hist.shape
```

```python
plt.hist(img.ravel(),256);
plt.show()
```

```python
img = cv2.imread('cat.jpg')
color = ('b','g','r')

for i , col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])
```

```python
mask = np.zeros(img.shape[:2],np.uint8)
mask[100:300,100:400] = 255
```

```python
img = cv2.imread('clahe.jpg',0);
```

```python
masked_img = cv2.bitwise_and(img,img,mask=mask)
```

```python
def cv_show(name,img):
    cv2.imshow('name',img);
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
cv_show('masked',masked_img)
```

```python
hist_full = cv2.calcHist([img],[0],None,[256],[0,256])
hist_mask = cv2.calcHist([img],[0],mask,[256],[0,256])
```

```python
plt.subplot(221),plt.imshow(img,'gray')
plt.subplot(222),plt.imshow(mask,'gray')
plt.subplot(223),plt.imshow(masked_img,'gray')
plt.subplot(224),plt.plot(hist_full),plt.imshow(img,'gray')
plt.xlim([0,256])
plt.show()
```

```python
equ = cv2.equalizeHist(img);
plt.hist(equ.ravel(),256)
plt.show()
```

```python
res = np.hstack((img,equ))
cv_show('res',res)
```

```python
clahe =cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))
res_clahe = clahe.apply(img)
res = np.hstack((img,equ,res_clahe))
cv_show('res',res)
```

```python
img = cv2.imread('lena.jpg',0)
img_float32 = np.float32(img)
dft = cv2.dft(img_float32,flags=cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)
mag_spec = 20*np.log(cv2.magnitude(dft_shift[:,:,0],dft_shift[:,:,1]))

plt.subplot(121),plt.imshow(img,cmap='gray')
plt.title('input image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(mag_spec,cmap='gray')
plt.title('mag spec'), plt.xticks([]),plt.yticks([])
plt.show()
```
