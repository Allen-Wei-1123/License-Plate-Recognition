```python
lena_noise = cv2.imread("lenaNoise.png")
```

```python
blur = cv2.blur(lena_noise,(3,3))
# cv2.imshow('blue',blur)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
box = cv2.boxFilter(lena_noise,-1,(3,3),normalize=False)
cv2.imshow('box',box)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gaussian = cv2.GaussianBlur(lena_noise,(5,5),1)
cv2.imshow('gaussian',gaussian)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
median = cv2.medianBlur(lena_noise,5);
cv2.imshow('median',median)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
res = np.hstack((blur,gaussian,median))
cv2.imshow('res',res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
