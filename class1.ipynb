```python
topsize,bottomsize,rightsize,leftsize = (50,50,50,50)

replicate = cv2.copyMakeBorder(newimg,topsize,bottomsize,rightsize,leftsize,borderType = cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(newimg,topsize,bottomsize,rightsize,leftsize,borderType = cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(newimg,topsize,bottomsize,rightsize,leftsize,borderType = cv2.BORDER_REFLECT_101)
constant = cv2.copyMakeBorder(newimg,topsize,bottomsize,rightsize,leftsize,borderType = cv2.BORDER_CONSTANT,value=1)
```

```python
plt.subplot(232), plt.imshow(replicate,'gray'), plt.title('replicate')
plt.subplot(233), plt.imshow(reflect,'gray'), plt.title('reflect')
plt.subplot(234), plt.imshow(reflect101,'gray'), plt.title('reflect101')
plt.subplot(235), plt.imshow(constant,'gray'), plt.title('constant')
plt.show()
```
